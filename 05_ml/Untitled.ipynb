```python
## 0.입력값 받기 - 첫 번째 줄
n, m, v = map(int, input().split())

## 1.딕셔너리에 키 값들 생성
graph = {}   # 그래프 노드와 간선 값을 저장할 딕셔너리 생성
for i in range(1, n+1):
    graph[i] = []
    
## 2.간선 정보를 입력받아 그래프 그리기
for _ in range(m):
    x, y = map(int, input().split())
    graph[x].append(y)
    graph[y].append(x)
print(graph)

## 3.DFS
visited = []      # 방문한 노드를 저장할 리스트
stack = [v]       # 방문 stack

# stack이 empty하면 while문 종료
while stack:
    node = stack.pop()

    # 만약 v가 방문하지 않은 노드라면, 실행
    # 이미 방문한 노드라면 아무 실행 없이 stack에서 빠지기만 함
    if node not in visited:
        visited.append(node)
        
        result = graph.get(node)
        result.sort(reverse=True)
        stack += result


## 4.BFS
visited2 = []
queue = [v]

# queue가 empty하면 while문 종료
while queue:
    node = queue.pop(0)
    
    # 만약 v가 방문하지 않은 노드라면, 실행
    # 이미 방문한 노드라면 아무 실행 없이 queue에서 빠지기만 함
    if node not in visited2:
        visited2.append(node)
        
        result = graph.get(node)
        result.sort()
        queue += result

print(' '.join(str(i) for i in visited), ' '.join(str(i) for i in visited2), sep='\n')
```
