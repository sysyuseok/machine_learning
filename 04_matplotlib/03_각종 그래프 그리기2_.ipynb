# 4. 파이차트 그리기
## 4.1 파이차트
- 각 범주(Category)가 데이터에서 차지하는 비율을 나타내는데 사용
- `pie(x, labels)` 이용
    - x: 값 (값들을 100을 기준으로 비율을 계산해 크기 설정)
    - labels : 값들의 label
    - autopct: 조각내에 표시될 비율의 문자열 형식. '%fmt문자' 
        - fmt문자: f(실수), d(정수), %% (%)
            

```python
"가격: %d, 길이: %.2f%%" % (100, 20.734)
```

```python
import matplotlib.pyplot as plt

labels = ['사과', '귤', '배', '바나나', '복숭아']
amounts = [20, 10, 50, 40, 90]

plt.figure(figsize=(8,8))
plt.pie(amounts, 
       labels=labels, 
       autopct='%.2f%%', 
       shadow=True, 
       explode=[0.2, 0, 0.2, 0, 0.7])

plt.show()
```

# 5 히스토그램(Histogram) 그리기
## 5.1 히스토그램

- 도수 분포표를 그래프로 나타낸 것.
    - 도수분포표: 연속형 자료를 특정 구간(bin)으로 나눠 그 빈도를 나타낸 표
        - 빈도 확인이나 **분포**를 볼때 사용한다.
    - X축: 계급(변수구간)의 대표값, Y축: 빈도수
- `hist(data [, bins=계급개수)` 메소드를 사용
    - data: 리스트형의 객체를 전달한다.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(100, 10, 1000)
data[:5]
# 평균 : 100, 표준편차: 10 의 정규분포를 따르는 난수 1000개 생성.
```

```python
s = pd.Series(data)
s.value_counts()
```

```python
pd.cut(s, bins=3).value_counts().to_frame().reset_index()
```

```python
v = pd.cut(s, bins=3).value_counts().to_frame()
v.reset_index(inplace=True)

plt.bar(v["index"].astype("str"), v["count"])
plt.show()
```

```python
# histogram
# 연속형 데이터 값들을 받아서 도수분포표를 계산한 뒤에 막대그래프로 수량을 시각화한다.
# default: 범위를 10등분
plt.hist(data, bins=50, edgecolor='black', color="white")
plt.show()
```

# 6. 상자그래프(Boxplot) 그리기
## 6.1 상자그래프란
- 연속형 값들의 사분위수, 극단값들을 이용해 값들의 분포를 시각화한다.
- boxplot(x, whis=1.5)
    - x: 데이터셋
    - whis: 극단치 계산시 사용할 값(기본 1.5)
        - IQR을 이용해 극단치 계산하기
            - IQR(Inter Quartile Range): $3분위수 - 1분위수$ 로 중간 50%값의 범위
            - 극단적으로 작은값: 1분위 - $IQR(3분위수-1분위수) * whis$   보다 작은 값들
            - 극단적으로 큰값: 3분위 + $IQR(3분위수-1분위수) * whis$  보다 큰 값들

```python
data = np.random.randint(1, 100, 1000)
data[:5] = [200, 250, -50, -100, -60]
```

```python
plt.figure(figsize=(3, 6))
plt.boxplot(data)
plt.show()
```

```python
plt.figure(figsize=(3, 6))
plt.boxplot(data, whis=4)  # IQR * 4 ==> 1.5 이상을 설정 ==> 정상 범위를 늘린다.
plt.show()
```

```python
plt.figure(figsize=(3, 6))
plt.boxplot(data, whis=0.2)  # IQR * 0.2 ==> 1.5 이하로 설정 ==> 정상 범위를 좁힌다.
plt.show()
```
